# Question.3

```python
using System;

namespace TriangleAreaCalculator
{
    class Question3
    {
        static void Main(string[] args)
        {
            double side1 = 2;
            double side2 = 2;
            double side3 = 2;

            double area = TriangleArea(side1, side2, side3);
            Console.WriteLine($"The area of a triangle with sides {side1}, {side2}, and {side3} is: {area}");
        }

        static double TriangleArea(double a, double b, double c)
        {
            double s = (a + b + c) / 2;
            double area = Math.Sqrt(s * (s - a) * (s - b) * (s - c));
            return area;
        }
    }
}
```

```python
using System;

namespace OddEvenSeparator
{
    class Question4
    {
        static void Main(string[] args)
        {
            // Ask user for the number of elements to be stored in the array
            Console.Write("Input the number of elements to be stored in the array: ");
            int numElements = Convert.ToInt32(Console.ReadLine());

            // Initialize arrays to store input, even, and odd numbers
            int[] inputArray = new int[numElements];
            int[] evenArray = new int[numElements];
            int[] oddArray = new int[numElements];
            int evenCount = 0, oddCount = 0;

            // Ask user to input elements into the array
            Console.WriteLine("Input {0} elements in the array:", numElements);
            for (int a = 0; a < numElements; a++)
            {
                Console.Write($"element - {a} : ");
                inputArray[a] = Convert.ToInt32(Console.ReadLine());

                // Separate even and odd numbers into their respective arrays
                if (inputArray[a] % 2 == 0)
                {
                    evenArray[evenCount] = inputArray[a];
                    evenCount++;
                }
                else
                {
                    oddArray[oddCount] = inputArray[a];
                    oddCount++;
                }
            }

            // Display even elements
            Console.WriteLine("The Even elements are:");
            for (int a = 0; a < evenCount; a++)
            {
                Console.Write(evenArray[a] + " ");
            }

            // Display odd elements
            Console.WriteLine("\nThe Odd elements are:");
            for (int a = 0; a < oddCount; a++)
            {
                Console.Write(oddArray[a] + " ");
            }
        }
    }
}
```

# Question.5

```python
using System;

namespace RectangleChecker
{
    class Question5
    {
        static void Main(string[] args)
        {
            // Test the inside() function with specific points and rectangles
            Console.WriteLine(inside(1, 1, 0, 0, 2, 3)); //  True
            Console.WriteLine(inside(-1, -1, 0, 0, 2, 3)); // False

            // Test the expression for both rectangles
            bool inFirstRectangle = inside(1, 1, 0.3, 0.5, 1.1, 0.7);
            bool inSecondRectangle = inside(1, 1, 0.5, 0.2, 1.1, 2);

            // Display whether the point (1,1) is inside each rectangle
            Console.WriteLine($"Is point (1,1) inside the first rectangle? {inFirstRectangle}");
            Console.WriteLine($"Is point (1,1) inside the second rectangle? {inSecondRectangle}");
        }

        // Check if a point (x, y) is inside a rectangle defined by two points (x1, y1) and (x2, y2)
        static bool inside(double x, double y, double x1, double y1, double x2, double y2)
        {
            // Check if (x, y) is within the rectangle's x and y boundaries
            return (x >= x1 && x <= x2) && (y >= y1 && y <= y2);
        }
    }
}
```

```python
def pig(word):
    vowels = ['a', 'e', 'i', 'o', 'u']
    word = word.lower()
    if word[0] in vowels:
        return word + 'way'
    else:
        return word[1:] + word[0] + 'ay'

print(pig('happy'))  
print(pig('Enter'))  
```

# Question.7

```python
def bldcount(filename):
    blood_count = {'A': 0, 'B': 0, 'AB': 0, 'O': 0, 'OO': 0}

    with open(filename, 'r') as file:
        content = file.read()
        blood_types = content.split()
        for blood_type in blood_types:
            blood_count[blood_type] += 1

    for blood_type, count in blood_count.items():
        if count == 0:
            print(f"There are no patients of blood type {blood_type}")
        elif count == 1:
            print(f"There is one patient of blood type {blood_type}.")
        else:
            print(f"There are {count} patients of blood type {blood_type}.")

# Call the function with the provided file path
bldcount(r"C:\Users\darsh\Downloads\bloodtype1.txt")
```

# Question.10

```python
def frequencies(text):
    letters = 'abcdefghijklmnopqrstuvwxyz'
    result = [0] * 26
    for char in text:
        if char.lower() in letters:
            index = letters.index(char.lower())
            result[index] += 1
    return result

# Testing the function with the given examples
print(frequencies('The quick red fox got bored and went home.'))
print(frequencies('apple'))
```
